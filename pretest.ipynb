```python
import pandas as pd
df_tr = pd.read_csv("train/train.tsv", delimiter='\t', header=None)
```

```python
df_tr['rule1'] = 0
df_tr['rule2'] = 0
df_tr['rule3'] = 0
```

```python
for i in range(0,16000):
    df_tr['rule1'][i] = df_tr[0][i].split('_')[0]
for i in range(0,16000):
    df_tr['rule2'][i] = df_tr[0][i].split('_')[1]
for i in range(0,16000):
    df_tr['rule3'][i] = df_tr[0][i].split('_')[2].split('.')[0]
```

```python
df_tr
```

```python
df = df_tr.sort_values(by=['rule1','rule2','rule3'])
```

```python
df.drop(['rule1','rule2','rule3'], axis = 'columns', inplace = True)
```
